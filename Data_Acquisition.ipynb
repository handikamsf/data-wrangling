```python
import requests
from bs4 import BeautifulSoup
```

```python
# Step 1: Fetch the Wikipedia page
url = "https://en.wikipedia.org/wiki/List_of_presidents_of_Indonesia"
response = requests.get(url)

# Step 2: Read the tables from the page
tables = pd.read_html(response.text)

# Step 3: Select the relevant table (assuming it's the third table)
tables2 = tables[2]

# Step 5: Extract unique lifespans
# Assuming the lifespan is in the first column (index 0)
unique_lifespans = tables2.iloc[:, 2].unique()

# Step 6: Convert to a list
unique_lifespans_list = unique_lifespans.tolist()

# Step 7: Create a new DataFrame from the unique lifespans
unique_lifespans_df = pd.DataFrame(unique_lifespans_list, columns=['Unique Lifespans'])

# Step 8: Display the new DataFrame
print(unique_lifespans_df)
```

```python
# Step 1: Send a request to the Wikipedia page
url = "https://en.wikipedia.org/wiki/Sukarno"
response = requests.get(url)

# Step 2: Check if the request was successful
if response.status_code == 200:
    # Step 3: Parse the HTML content
    soup = BeautifulSoup(response.text, 'html.parser')

    # Step 4: Find the infobox containing the biodata
    infobox = soup.find("table", {"class": "infobox"})

    # Step 5: Extract data from the infobox
    biodata1 = {}
    for row in infobox.find_all("tr"):
        header = row.find(['th', 'span'])
        data = row.find("td")
        if header and data:
            biodata1[header.text.strip()] = data.text.strip()

    # Step 6: Print the extracted biodata
    for key, value in biodata1.items():
        print(f"{key}: {value}")
else:
    print("Failed to retrieve the page")
```

```python
# prompt: connect psql to python

!pip install psycopg2-binary

import psycopg2

def connect_to_psql(host, database, user, password, port=5432):
  """Connects to a PostgreSQL database.

  Args:
    host: The hostname or IP address of the database server.
    database: The name of the database to connect to.
    user: The username to use for authentication.
    password: The password to use for authentication.
    port: The port number the database server is listening on (default: 5432).

  Returns:
    A psycopg2 connection object if successful, None otherwise.
  """
  try:
    conn = psycopg2.connect(
        host=host,
        database=database,
        user=user,
        password=password,
        port=port
    )
    print("Connected to PostgreSQL successfully")
    return conn
  except psycopg2.Error as e:
    print(f"Error connecting to PostgreSQL: {e}")
    return None


# Example usage (replace with your database credentials)
# host = 'your_host'
# database = 'your_database'
# user = 'your_user'
# password = 'your_password'

# conn = connect_to_psql(host, database, user, password)

# if conn:
#     # Perform database operations here
#     cur = conn.cursor()
#     # Example: Execute a query
#     cur.execute("SELECT version()")
#     db_version = cur.fetchone()
#     print(f"PostgreSQL database version: {db_version}")
#     cur.close()
#     conn.close()
```
